```python
import pandas as pd
df = pd.read_csv("./CSV/DecisionTree.csv")
df.head()
```

```python
inputs = df.drop('Drug',axis = 1)
target = df['Drug']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
target = le.fit_transform(target)
target
```

```python
inputs['Gender'] = le.fit_transform(inputs['Gender'])
inputs['BP'] = le.fit_transform(inputs['BP'])
inputs['Cholesterol'] = le.fit_transform(inputs['Cholesterol'])
inputs
```

```python
from sklearn import tree
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,target_train,target_test = train_test_split(inputs,target,test_size = 0.1)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(inputs,target)
```

```python
model.score(input_test,target_test)
```
