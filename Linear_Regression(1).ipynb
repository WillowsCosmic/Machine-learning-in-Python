```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./CSV/SimpleLinear.csv")
df
```

```python
%matplotlib inline
plt.xlabel('years_experience');
plt.ylabel('salary');
plt.scatter(df.years_experience,df.salary);
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['years_experience']],df.salary) #salary is y,years is x
```

```python
reg.predict([[7.2]])
```
