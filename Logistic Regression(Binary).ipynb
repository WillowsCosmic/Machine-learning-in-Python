```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./CSV/Binary Classification.csv")
df
```

```python
plt.scatter(df.Age,df.Insurance_Bought)
plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df[['Age']],df.Insurance_Bought,test_size = 0.1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
